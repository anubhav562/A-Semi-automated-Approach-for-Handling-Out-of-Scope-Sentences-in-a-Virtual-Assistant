```python
import json
import pandas as pd
```

```python
# we will read the JSON data here
with open("data_full.json") as f:
    json_data_dict = json.load(f)
```

```python
# viewing what all keys do we have in the data
json_data_dict.keys()
```

```python
# we will choose the train key and will select a subset of data from train set
data_list = json_data_dict.get("train")
data_list_2 = json_data_dict.get("val")
data_list_3 = json_data_dict.get("test")

final_data_list = []

for data in data_list:
    final_data_list.append(data)
    
for data in data_list_2:
    final_data_list.append(data)
    
for data in data_list_3:
    final_data_list.append(data)
```

```python
# viewing all the intents that we have in this subset of data
intent_list = list(set([utterance[1] for utterance in final_data_list]))
```

```python
intent_list
```

```python
# we will select a few intents (10-15) from the above list and make the final dataset

intent_to_sentence_mapping = {}

for utterance in final_data_list:
    if utterance[1] in intent_to_sentence_mapping.keys():
        intent_to_sentence_mapping[utterance[1]].append(utterance[0])
    else:
        intent_to_sentence_mapping[utterance[1]] = [utterance[0]]
```

```python
# select a random subset of intents for our experiments.

selected_intents = [
    "transfer", "timer", "change_language", "time", "application_status", "flight_status", "balance",
    "traffic", "calories", "shopping_list", "make_call", "alarm", "find_phone", "next_holiday", "weather"
]
```

```python
# form a dictionary/mapping for the same

selected_intents_mapping = {}
for intent in selected_intents:
    selected_intents_mapping[intent] = intent_to_sentence_mapping[intent]
```

```python
selected_intents_mapping
```

```python
# unpacking the sentences in a flat representation and making a dataframe
flat_sentences_list = []
for intent in selected_intents:
    for utterance in selected_intents_mapping[intent]:
        flat_sentences_list.append(utterance)
        
utterance_df = pd.DataFrame(flat_sentences_list, columns=["utterance"])
utterance_df
```

```python
# exporting data 

utterance_df.to_csv("utterance_data.csv", index=False)
```
